```python
import pandas as pd
```

```python
dataset = pd.read_csv("2020-2021_Assessment_Roll.csv")
dataset
```

```python
columns = ["Print Key", "Previous Property Class", "Owner2", "Mail1", "Mail2", "Zipcode Extension", "Year Built", "Overall Condition", "Building Style", "Heat Type", "Basement Type", "# of Stories", "# of Fireplaces", "# of Beds", "# of Baths", "# of Kitchens","Location"]
dataset = dataset.drop(columns = columns)
```

```python
dataset = dataset.dropna()
```

```python
dataset.shape
```

```python
# Checking for null values in the dataset
null_values = dataset.isnull().sum()

# Displaying columns with null values (if any)
print(null_values[null_values > 0])
```

```python
dataset.dtypes
```

```python
dataset["Zipcode"] = dataset["Zipcode"].astype(str)
dataset["Index"] = range(1, len(dataset) + 1)
dataset = dataset[:25000]
```

```python
dataset.shape
```

```python
# Table 1: OWNER info
owner_info = dataset[["Owner1", "Previous Owner", "Mail3", "Mail4"]]

# Table 2: House info
address_info = dataset[["House Number", "Street", "Zipcode","SBL"]]

#Table 3: Zipcode table
zipcode_info = dataset[["Zipcode","State","City"]]

# Table 4: Deed info
deed_info = dataset[["Deed Book", "Deed Page", "Deed Date", "Roll"]]

#table 5:Census table
census_info =  dataset[["Census Tract"]]

# Table 6: Property details
property_info = dataset[["SBL", "Front", "Depth", "Property Class", "Total Living Area"]]

#Table 7: Property class table
property_class_info = dataset[["Property Class","Property Class Description"]]

# Table 8: Price details
price_info = dataset[["Land Value", "Total Value", "Sale Price"]]

# Generating surrogate primary keys
owner_info['Owner_ID'] = dataset["Index"]
deed_info['Deed_ID'] = dataset["Index"]
price_info['Price_ID'] = dataset["Index"]
census_info['Census_ID'] = dataset["Index"]

#Table 9: link table
link_table = pd.DataFrame({
    "SBL": property_info["SBL"],
    "Owner_ID": owner_info["Owner_ID"],
    "Deed_ID": deed_info["Deed_ID"],
    "Price_ID": price_info["Price_ID"],
    "Census_ID": census_info["Census_ID"]
})
```

```python
# Save owner_info to CSV
owner_info.to_csv('owner_info.csv', index=False)

# Save address_info to CSV
address_info.to_csv('address_info.csv', index=False)

# Save zipcode_info to CSV
zipcode_info.to_csv('zipcode_info.csv', index=False)

# Save deed_info to CSV
deed_info.to_csv('deed_info.csv', index=False)

# Save census_info to CSV
census_info.to_csv('census_info.csv', index=False)

# Save property_info to CSV
property_info.to_csv('property_info.csv', index=False)

# Save property_class_info to CSV
property_class_info.to_csv('property_class_info.csv', index=False)

# Save price_info to CSV
price_info.to_csv('price_info.csv', index=False)

# Save link_table to CSV
link_table.to_csv('link_table.csv', index=False)
```
